```python
handleGen = Handle('std::vector<reco::GenParticle>')
labelGen = ("genParticles")
handleSA = Handle('std::vector<reco::Track>')
labelSA = "standAloneMuons"
handleRSA = Handle('std::vector<reco::Track>')
labelRSA = "refittedStandAloneMuons"
```

```python
genMuonPt = ROOT.TH1F("genMuonPt", "genMuon p_T", 400, 0, 400)
genMuonEta = ROOT.TH1F("genMuonEta", "genMuon \eta", 100, -2.4, 2.4)
muonSAPtAll = ROOT.TH1F("muonSAPtAll", "standAloneMuon p_T before cuts", 400, 0, 400)
muonSAPt = ROOT.TH1F("muonSAPt", "standAloneMuon p_T after cuts", 400, 0, 400)
muonRSAPtAll = ROOT.TH1F("muonRSAPtAll", "refittedStandAloneMouns p_T before cuts" , 400, 0, 400)
muonRSAPt = ROOT.TH1F("muonRSAPt", "refittedStandAloneMouns p_T after cuts", 400, 0, 400)

xMin = -1
xMax = 1
# xMin = -1
# xMax = 1
# Barrel: |eta| < 0.8
muonSACurvResolBarrel = ROOT.TH1F("muonSACurvResolBarrel", "standAloneMuon q/p_T resolution for |eta| < 0.8", 100, xMin, xMax)
muonRSACurvResolBarrel = ROOT.TH1F("muonRSACurvResolBarrel", "refittedStandAloneMuon q/p_T resolution for |eta| < 0.8", 100, xMin, xMax)
# Overlap: 0.8 <= |eta| < 1.2
muonSACurvResolOverlap = ROOT.TH1F("muonSACurvResolOverlap", "standAloneMuon q/p_T resolution for 0.8 <= |eta| < 1.2", 100, xMin, xMax)
muonRSACurvResolOverlap = ROOT.TH1F("muonRSACurvResolOverlap", "refittedStandAloneMuon q/p_T resolution for 0.8 <= |eta| < 1.2", 100, xMin, xMax)
# Endcaps: 1.2 <= |eta| < 2.4
muonSACurvResolEndcaps = ROOT.TH1F("muonSACurvResolEndcaps", "standAloneMuon q/p_T resolution for 1.2 <= |eta| < 2.4", 100, xMin, xMax)
muonRSACurvResolEndcaps = ROOT.TH1F("muonRSACurvResolEndcaps", "refittedStandAloneMuon q/p_T resolution for 1.2 <= |eta| < 2.4", 100, xMin, xMax)

muonSAD0Resol = ROOT.TH1F("muonSAD0Resol", "standAloneMuons d0 resolution", 100, -10, 10)
muonRSAD0Resol = ROOT.TH1F("muonRSAD0Resol", "refittedStandAloneMuons d0 resolution", 100, -10, 10)

# Reconstruction efficiency as a function of d0 and z0
genMuonD0 = ROOT.TH1F("genMuonD0", "generated muon d0", 20, 0, 100)
matchedMuonSAD0 = ROOT.TH1F("matchedMuonSAD0", "matched standAloneMuon d0", 20, 0, 100)
matchedMuonRSAD0 = ROOT.TH1F("matchedMuonRSAD0", "matched refittedStandAloneMuon d0", 20, 0, 100)
```

```python
def fillPlots(genMuons, muons, muonPtAll, muonPt, muonCurvResolBarrel, muonCurvResolOverlap, muonCurvResolEndcaps, muonD0Resol):
    for muon in muons:
        muonPtAll.Fill(muon.pt())
        if (muon.hitPattern().numberOfValidMuonDTHits() + muon.hitPattern().numberOfValidMuonCSCHits() > 24) \
        and (muon.hitPattern().dtStationsWithValidHits() + muon.hitPattern().cscStationsWithValidHits() > 2):
            muonPt.Fill(muon.pt())
            for genMuon in genMuons:
                if deltaR(genMuon, muon) < 0.2:
                    resol = (muon.charge()/muon.p() - genMuon.charge()/genMuon.p()) / (genMuon.charge()/genMuon.p())
                    # resol = (muon.p() - genMuon.p()) / genMuon.p()
                    resolD0 = muon.d0() - genMuon.vertex().x()
                    # print "gen d0 =", genMuon.vertex().x()
                    # print "reco d0=", muon.d0() 
                    # print "gen pt =", genMuon.pt()
                    # print "reco pt=", muon.pt()
                    if abs(genMuon.vertex().x()) > 1:
                        resolD0 = (muon.d0() - genMuon.vertex().x())/genMuon.vertex().x()
                    muonD0Resol.Fill(resolD0)
                    if abs(muon.eta()) < 0.8:
                        muonCurvResolBarrel.Fill(resol)
                    elif abs(muon.eta()) < 1.2:
                        muonCurvResolOverlap.Fill(resol)
                    elif abs(muon.eta()) < 2.4:
                        muonCurvResolEndcaps.Fill(resol)
    
# Reconstruction efficiency - THIS NEEDS TO BE IMPROVED OR MOVED TO ANOTHER NOTEBOOK PAGE
def fillRecoEffPlots(genMuons, muons, genHisto, recoHisto):
    for genMuon in genMuons:
        if isinstance(genHisto, ROOT.TH1F): genHisto.Fill(abs(genMuon.vertex().x()))
        for muon in muons:
            if (muon.hitPattern().numberOfValidMuonDTHits() + muon.hitPattern().numberOfValidMuonCSCHits() > 24) \
            and (muon.hitPattern().dtStationsWithValidHits() + muon.hitPattern().cscStationsWithValidHits() > 2):
                if deltaR(genMuon, muon) < 0.2:
                    recoHisto.Fill(abs(genMuon.vertex().x()))
```

```python
# events = Events("/afs/cern.ch/user/d/demattia/work/MuonGun_jun_10_2013/muongun_pt_100_GeV_d0_uniform_0_100cm_7000events.root")
# events = Events("/afs/cern.ch/user/d/demattia/work/MuonGun_jun_7_2013/muongun_pt_100_GeV_d0_prompt_7000events.root")
events = Events("/afs/cern.ch/user/d/demattia/work/MuonGun_jun_18_2013_pt_100GeV_d0_300cm/muongun_pt_100GeV_d0_300cm_7000events.root")
for event in events:
    
    event.getByLabel(labelGen, handleGen)
    genMuons = handleGen.product()
    # event.getByLabel(labelSA, "UpdatedAtVtx", handleSA)
    event.getByLabel(labelSA, handleSA)
    muonsSA = handleSA.product()
    event.getByLabel(labelRSA, handleRSA)
    muonsRSA = handleRSA.product()

    for genMuon in genMuons:
        genMuonPt.Fill(genMuon.pt())
        genMuonEta.Fill(genMuon.eta())

    fillPlots(genMuons, muonsSA, muonSAPtAll, muonSAPt, muonSACurvResolBarrel, muonSACurvResolOverlap, muonSACurvResolEndcaps, muonSAD0Resol)
    fillPlots(genMuons, muonsRSA, muonRSAPtAll, muonRSAPt, muonRSACurvResolBarrel, muonRSACurvResolOverlap, muonRSACurvResolEndcaps, muonRSAD0Resol)
    
    fillRecoEffPlots(genMuons, muonsSA, genMuonD0, matchedMuonSAD0)
    fillRecoEffPlots(genMuons, muonsSA, 0, matchedMuonSAD0)
```

Define utility functions (should be moved to a utility module)

```python
def roundToN(value, roundValue, n=0):
    return round(value, -int(math.floor(math.log10(roundValue)))+1)
def printValueError(value, error):
    return str(roundToN(value, error)) + " +/- " +str(roundToN(error, error))
def drawResol(resolHisto):
    resolHisto.Draw()
    resolHisto.Fit("gaus")
    func = resolHisto.GetFunction("gaus")
    print resolHisto.GetTitle()
    print "chi2 = %.2f" %func.GetChisquare(), "/", func.GetNDF()
    print "mean =", printValueError(func.GetParameter(1), func.GetParError(1))
    print "sigma =", printValueError(func.GetParameter(2), func.GetParError(2))
```
